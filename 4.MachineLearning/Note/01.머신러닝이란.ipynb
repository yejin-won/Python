# 머신러닝이란?
: 수 많은 데이터를 학습시켜 거기에 있는 패턴(y=ax+b)을 찾아내고 이를 기반으로 데이터를 분류하거나 예측.
- 일반 소프트웨어(결정론적 소프트웨어)는 언제나 똑같은 과정을 거쳐 똑같은 결과를 냅니다.
- 머신러닝(비결정론적 소프트웨어) 기반 소프트웨어는 데이터와 알고리즘(예. ctree, randomForest..)에 따라 때로는 다른 결과를 추론해 냅니다. 
: 추론 : 이미 알고 있는 것으로부터 논리적 결론을 도출하는 과정

### 결정론적 소프트웨어와 비결정론적 소프트웨어의 차이점
: 예를 들어 지문인식기의 소프트웨어를 결정론적 소프트웨어로 처리하였다고 하면 지문 입력기에 지문이 조금만 틀어져도 거부 메세지가 나온다. 이에 반해 비결정론적 소프트웨어는 지문이 조금 틀어져도 추론을 통해 지문의 가부를 결정합니다.

### 머신러닝 개요
: 머신러닝이란 인공지능 연구 과제 중의 하나로 인간의 뇌가 자연스럽게 수행하는 '학습'이라는 능력을 컴퓨터로 구현하는 방법      
- 어느 정도 규모가 있는 샘플 데이터를 입력으로 넣어 분석 시킵니다.
- 그리고 이때 데이터에서 일정한 규칙을 찾아 냅니다.     
- 그리고 찾아낸 규칙을 기반으로 다른 데이터로 분류(classfication)하거나 예측(regression)하는 것.      

### 머신러닝의 종류
##### 1) 지도학습(Supervised Learning)
: 학습시 데이터와 답을 같이 학습한 후 다른 데이터의 값을 예측 (데이터가 무조건 많아야 함)
딥러닝과의 차이점 => 분석가가 관여하므로 결과가 달라짐

##### 2) 비지도학습(Unsupervised Learning)
: 학습시 데이터만 학습한 후 (답이 없으므로) 다른 데이터의 규칙성을 찾습니다. 본질적인 구조 등을 확인할 때 사용
예) 데이터가 너무 많거나 컬럼이 너무 많은 경우, 줄이기 위해 사용
##### 3) 강화학습(Reinforcement Learning)
: 부분적으로 답을 입력한 후 데이터를 기반으로 최적의 답을 찾아 냅니다. (수많은 시행착오를 거쳐 답을 찾아냄)

##### 4) 딥러닝(Deep Learning)
: 지도학습과 비지도학습의 관련이 있으나 특이성은 학습자료에 대한 인간의 개입을 받지 않는다.

(1) 데이터 수집 <br>
(2) 데이터 가공 <br>
(3) 데이터 학습 <br>
    &nbsp; &nbsp; &nbsp;(3-1) 학습 방법 선택 <br>
    &nbsp; &nbsp; &nbsp;(3-2) 매개 변수 조정 <br>
    &nbsp; &nbsp; &nbsp;(3-3) 모델 학습 <br>

(5) 예측력이 좋은 경우 다음 단계 진행하고 아니면 (3)번부터 다시 <br>
(6) 성공 

# 머신러닝 응용분야
### 클래스 분류(Classification)
- 특정 데이터에 레이블(label=>정답 값)을 붙여 분류할 수 있습니다.(label없으면 비지도학습)
- 스팸 메일 분류, 필기인식, 증권 사기 방지 등

### 클러스터링(Clustering)
- 값의 유사성을 기반으로 데이터를 여러 그룹으로 나눌 수 있습니다.
- 사용자의 취향을 그룹으로 묶어 사용자 취향에 맞는 광고를 제공

### 추천(Recommendation)
- 특정 데이터를 기반으로 다른 데이터를 추천하는 것입니다.
- 인터넷 서점에서 구매한 책들을 기반으로 다른 책을 추천.

### 회귀(Regression)
- 과거의 데이터를 기반으로 미래를 예측한 것입니다.
- 판매 예측, 주가 변동

### 차원 축소(Dimensionality Reduction)
- 데이터의 특성을 유지하면서 데이터의 양을 줄이는 것입니다.
- 데이터를 시각화 하거나 구조를 추출해서 용량을 줄여 계산을 빠르게 하거나 메모리를 절약할 때 사용

# 프로젝트 과정
: 머신러닝 프로젝트는 크게 문제정의, 데이터 획득, 모델 구현, 검증, 테스트의 단계로 구현

### 문제 정의
- 프로젝트의 목표를 정의하고 그에 따른 계획을 세우는 단계
- 대략적으로 어떤 데이터를 수집해서 어떤 머신러닝 알고리즘으로 문제를 해결할지 결정하는 단계

### 데이터 획득
- 머신러닝 모델 학습에 필요한 데이터를 수집하는 단계
- 성공적인 머신러닝 프로젝트는 뛰어난 머신러닝 실력보다도 얼마나 충분히 데이터를 획득했느냐에 의해 결정이 됩니다.
- 획득한 데이터는 학습데이터, 검증데이터, 테스트 데이터로 구분되어 사용 됩니다.

####### 데이터 양에 따라 다르지만 많은 경우(몇십만이상), test 20% 나머지 80% 트레이닝으로 사용할수도...
####### 20%를 일반화 성능으로 하고 다시 80%에서 20% test(조절) 60% 트레이닝으로 나눠 성능을 테스트 

### 모델 구현
- 학습 데이터를 기반으로 한 한개 이상의 머신러닝 모델을 구현

### 검증
- 검증 데이터를 사용해 구현된 머신러닝 모델들의 성능을 검증
- 최종적으로 검증 결과가 가장 높은 모델 하나를 선택

### 테스트
- 검증 결과를 통해 선택된 모델의 성능을 테스트 데이털로 측정(일반화 성능 측정)
