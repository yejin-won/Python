### Pandas 내장 그래프 도구

```python
df = pd.read_excel("../Data/남북한발전전력량.xlsx")
df
```

```python
df_sn = pd.DataFrame(df.iloc[[0,5],2:])
df_sn.index = ['South','North']
df_sn
```

```python
df_sn = pd.DataFrame(df.loc[[0,5],'1990':])
df_sn.index = ['South','North']
df_sn
```

```python
df_sn = pd.DataFrame(df.loc[[0,5],'1990':])
df_sn.index = ['South','North']
df_sn.columns = df_sn.columns.map(int) # 열 이름의 자료형을 정수형으로 변경(차트그릴때 유용)
```

```python
df_sn = df_sn.loc[:,1990:]
df_sn
```

```python
# 선 그래프 그리기

tdf_sn = df_sn.transpose()
tdf_sn
```

```python
tdf_sn.plot()
```

```python
# 막대 그래프 그리기

tdf_sn.plot(kind='bar')
```

```python
# 히스토그램
tdf_sn.plot(kind='hist')
```
