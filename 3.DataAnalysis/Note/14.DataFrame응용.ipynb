### 함수 Mapping
: Series 또는 DataFrame의 Data를 특정함수에 일대일 대응 시키는 과정


```python
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# titanic 데이터셋에서 age, fare 2개 열을 선택하여 데이터프레임 만들기
df = titanic.loc[:,['age','fare']]
df['ten'] = 10
df.head()
```

```python
# 사용자 함수 정의 
def add_10(n): # 10을 더하는 함수
    return n + 10
def add_two_obj(a,b):
    return a + b
```

```python
print(add_10(10))
print(add_two_obj(10,10))
```

```python
# 시리즈 객체에 적용
sr1 = df['age'].apply(add_10) # df['age'] = n 의미
sr1.head()
```

```python
# 시리즈 객체에 숫자 적용
sr2 = df['age'].apply(add_two_obj,b=10) # age + 10
sr2.head()
```

```python
# 람다함수 :  시리즈 객체에 적용
sr3 = df['age'].apply(lambda x: x + 10) # 단점 : 계산식이 복잡해지면 사용 어려움
sr3.head() 
```

```python
sr4 = df['age'].apply(lambda x: add_10(x)) # 사용자 정의 함수 x값을 넘겨서 계산 출력
sr4.head() 
```

```python
# DataFrame에 함수 Mapping
df_map =  df.applymap(add_10) # 모든 컬럼에 10을 더함 
df_map.head()
```

```python
def missing_value(series):
    return series.isnull()
```

```python
df.apply(missing_value) # apply 하나의 컬럼(시리즈) 당 체크, applymap 한 번에 체크
```

> apply는 column별(series)로 함수를 실행하나, applymap은 dataframe 전체로 함수 실행

```python
# 데이터프레임의 각 열의 최대값과 최소값의 차이를 출력
def min_max(series):
    return series.max() - series.min()
```

```python
result = df.apply(min_max)
print(result)
```

```python
# lamda 사용
df['add'] = df.apply(lambda x : add_two_obj(x['age'],x['ten']),axis=1)
df.head()
```

```python
df.tail()
```

### pipe() 사용하기

```python
df = titanic.loc[:,['age','fare']]
df.head()
```

```python
# 각 열의 NaN 찾기
def missing_value(x):
    return x.isnull()
```

```python
# 각 열의 NaN 갯수 변환
def missing_count(x):
    return missing_value(x).sum()
```

```python
# 총 NaN 갯수
def total_number_missing(x):
    return missing_count(x).sum()
```

```python
df.pipe(missing_value)
```

```python
df.pipe(missing_count)
```

```python
df.pipe(total_number_missing)
```

### Filtering
- 특정 조건식을 만족하는 Data만 따로 추출하는 방법

```python
# 나이가 10대(10~19세)인 승객만 따로 선택
df_teenage = titanic[(titanic.age >=10) & (titanic.age <20)]
df_teenage.head()
```

```python
# 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택

df_female_under10 = titanic[titanic.age < 10]
df_female_under10.head()
```

```python
# 나이가 10세미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택

df_under10_morethan60 = titanic[(titanic.age <10) | (titanic.age >=60)].loc[:,['age','sex','alone']]
df_under10_morethan60.head()
```

### DataFrame 병합
- SQL의 Join과 비슷한 방식으로 어떤 기준에 의해 병합하는 개념
- 이때 기준이 되는 열이나 index를 Key라고 한다.
- Key가 되는 열이나 index는 반드시 양쪽 DataFrame에 존재 해야한다.

```python
# 주식 데이터를 가져와서 데이터 프레임 만들기
df1 = pd.read_excel("../Data/stock price.xlsx")
df1.head()
```

> id : 종목 코드     
> stock_name : 회사 이름     
> value : 시가 총액      
> price : 주가    

```python
df2 = pd.read_excel("../Data/stock valuation.xlsx")
df2.head()
```

> id : 종목 코드     
> name : 회사명       
> eps : 주당순이익         
> bps : 주당 순자산가치      
> per : 주가 수익비율   
> pbr : 주가 자산비율           

```python
# 데이터프레임 합치기 - 교집합
merge_inner = df1.merge(df2,how='inner')
# merge_inner = pd.merge(df1,df2,how='inner',on='id')
merge_inner.head()
```

```python
# 데이터 프레임 합치기 - 합집합
merge_outer = df1.merge(df2,how='outer')
# merge_outer = pd.merge(df1,df2,how='outer',on='id')

merge_outer.head()
```

```python
# 데이터프레임 합치기 - 왼쪽 데이터프레임 기준, 기준값은 회사명
merge_left = pd.merge(df1,df2,how='left',left_on='stock_name',right_on='name')
merge_left.head()
```

```python
# 데이터프레임 합치기 - 오른쪽 데이터프레임 기준, 기준값은 회사명
merge_right = pd.merge(df1,df2,how='right',left_on='stock_name',right_on='name')
merge_right.head()
```

```python
# df1에서 price가 50000미만인 데이터 찾기
price = df1[df1.price < 50000]
price.head()
```

```python
value = pd.merge(price,df2)
value
```

```python
value.drop('name',axis=1,inplace=True)
value
```

# 그룹 연산
- 복잡한 데이터를 어떤 기준에 따라 여러 그룹으로 나눠서 관찰하는 것도 좋은 방법이다.
- 특정 기준을 적용하여 몇개의 그룹으로 분할하여 처리하는 과정이 그룹 연산이다.
- 1단계 : 분할(split) -> 데이터를 특정 조건에 의해 분할
- 2단계 : 적용(apply) -> 데이터를 집계, 변환, 필터링
- 3단계 : 결합(combine) -> 2단계 처리 결과를 하나로 결합

### 그룹 객체 만들기 (분할 단계)

```python
titanic.head()
```

```python
df = titanic.loc[:,['age','sex','class','fare','survived']]
df.head()
```

```python
# class열을 기준으로 분할 
df['class'].unique() # category(범주형)으로 되어있어야 데이터를 분리할 수 있음
```

```python
grouped = df.groupby(['class']) # [(First,Data1),(Second,Data2),(Third,Data3)] 리스트 안에 튜플들 있음
list(grouped)
```

```python
# 그룹객체를 iteration으로 출력 : 처음부터 5줄만 출력
for key, group in grouped:
    print("* key :",key)
    print("* number :" ,len(group))
    print(group.head())
    print("="*60)
```

```python
g3 = grouped.get_group('Third').head()
type(g3)
```

```python
# 연산 메소드 적용
grouped.mean()
```

```python
# class와 sex열을 기준으로 분할
df['sex'].unique() # object이므로 category 시켜야함 (but 3.5이상에서는 하지 않아도 가능)
```

```python
df.info()
```

```python
grouped_sex = df.groupby(['sex'])
list(grouped_sex)
```

```python
# class와 sex열을 기준으로 분할
grouped_two = df.groupby(['class','sex'])
list(grouped_two)
```

```python
# 그룹객체를 iteration으로 출력 : 처음부터 5줄만 출력
for key, group in grouped_two:
    print("* key :",key)
    print("* number :" ,len(group))
    print(group.head())
    print("="*60)
```

```python
grouped_two.mean()
```

```python
# grouped_two에서 Thrid의 female만 group3f라는 dataframe 만들기

grouped_two.keys
```

```python
group3f = grouped_two.get_group(('Third','female'))
group3f.head()
```
