# Matplotlib
: 파이썬의 표준 시각화 도구

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../Data/시도별 전출입 인구수.xlsx")
df.head()
```

```python
df.tail()
```

```python
df['전출지별'].isnull()
df['전입지별'].isnull()
```

```python
# NaN을 전(prev) 데이터로 채움
df = df.fillna(method='ffill')
df.head()
```

```python
df.info()
```

```python
df
```

```python
# 서울에서 다른 지역으로 이동한 데이터만 추출하여 정리
# 전출지별이 서울특별시 이고 전입지별이 서울특별시가 아닌 곳
df_seoul = df[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
df_seoul
```

```python
# 전출지별 컬럼 삭제
df_seoul.drop(columns='전출지별',inplace=True)
```

```python
df_seoul
```

```python
# 전입지별 ---> 전입지로 컬럼이름 변경

df_seoul.rename({'전입지별' : '전입지'},axis=1,inplace=True)
```

```python
df_seoul[(df_seoul['전입지'] == '전국') | (df_seoul['전입지'] == '대구광역시') | (df_seoul['전입지'] == '부산광역시')]
```

```python
# df_seoul의 index를 전입지로 변경
df_seoul.set_index('전입지',inplace=True)
df_seoul.head()
```

```python
# 서울에서 경기도로 이동한 인구 데이터 값만 선택
sr_one = df_seoul.loc['경기도']
sr_one.head()
```

```python
plt.plot(sr_one.index,sr_one.values)
plt.show()
```

```python
plt.plot(sr_one) # 데이터의 수가 적으므로 가능 
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 차트 제목 추가
plt.plot(sr_one)
plt.title("서울 => 경기 인구 이동")

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.show()
```

```python
# 그래프 꾸미기
# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))

# x축 눈금 라벨 회전하기
plt.xticks(rotation=45)

# 차트 제목 추가
plt.title("서울 => 경기 인구 이동")

# y축 눈금 라벨 회전 하기
plt.yticks(rotation=45)

# 축 이름 추가
plt.xlabel('기간')
plt.ylabel('이동 인구수')
plt.plot(sr_one)

# 범례 표시
plt.legend(labels = ['서울->경기'],loc='best')
plt.show()
```

### Style 추가해보기

```python
# 그래프 꾸미기

# Style 서식 지정
plt.style.use('ggplot')

# 그림 사이즈 지정(가로 14인치, 세로 5인치)
plt.figure(figsize=(14,5))

# x축 눈금 라벨 회전하기
plt.xticks(rotation=45)

# 차트 제목 추가
plt.title("서울 -> 경기 인구 이동",size=20)

# y축 눈금 라벨 회전 하기
plt.yticks(rotation=45)

# 축 이름 추가
plt.xlabel('기간',size=20)
plt.ylabel('이동 인구수',size=20)

# Marker 표시
plt.plot(sr_one,marker='o',markersize=10)

# y축 범위 지정
plt.ylim(50000,800000) # 최소값, 최대값

# 화살표 표시
plt.annotate(
    '',
    xy = (20,620000), # 화살표의 머리부분
    xytext = (2,290000), # 화살표의 꼬리부분
    xycoords = 'data', # 좌표체계 3.7이상부터는 안써도 됨
    arrowprops = dict(arrowstyle='->',color='skyblue',lw=5), 
)
plt.annotate('인구 이동 증가(1970~1995)', # 텍스트
             xy= (10, 400000), # 텍스트 위치 기준점
             rotation =28, # 텍스트 회전 각도
             va = 'baseline', # 텍스트 좌우 정렬
             fontsize = 12 # 텍스트 크기
)

plt.annotate('',
             xy= (45,400000),
             xytext = (30,600000),
             xycoords = 'data',
             arrowprops = dict(arrowstyle='->',color='olive',lw=5)

)
plt.annotate('인구 이동 감소(1995~2017)',
             xy = (30,500000),
             rotation =-17, # 텍스트 회전 각도
             va = 'baseline', # 텍스트 좌우 정렬
             fontsize = 12 # 텍스트 크기
)

# 범례 표시
plt.legend(labels = ['서울->경기'],loc='best',fontsize = 15)

# 이미지 저장
plt.savefig("../Data/seoul.png")

plt.show()
```

## 동일한 그림에 여러개의 그래프를 추가

```python
# 서울에서 '충청남도, '경상북도', '강원도' 로이동한 데이텀만 추출하여 정리
df_3 = df_seoul.loc[['충청남도','경상북도','강원도']]
df_3
```

```python
# 서울에서 '충청남도, '경상북도', '강원도' 로이동한 데이텀만 추출하여 정리
df_3 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도']]
df_3
```

```python
# Style
plt.style.use('ggplot')

fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(1,1,1)
ax.plot(df_3.columns,
        df_3.loc['충청남도'],
        marker = 'o',
        markerfacecolor = 'green',
        markersize = 10,
        color = 'olive',
        linewidth = 2,
        label = '서울->충남'
)
ax.plot(df_3.columns,
        df_3.loc['경상북도'],
        marker = 'o',
        markerfacecolor = 'blue',
        markersize = 10,
        color = 'skyblue',
        linewidth = 2,
        label = '서울->경북'
)
ax.plot(df_3.columns,
        df_3.loc['강원도'],
        marker = 'o',
        markerfacecolor = 'red',
        markersize = 10,
        color = 'magenta',
        linewidth = 2,
        label = '서울->강원'
)


plt.legend(loc='best')
ax.set_title('서울->충남,경북,강원 인구 이동',size=20)

ax.set_xlabel('기간',size=12)
ax.set_ylabel('이동 인구수',size=12)

ax.set_xticklabels(df_3.columns,rotation=90)

# 축 눈금 라벨 크기
ax.tick_params(axis='x',labelsize=5)
ax.tick_params(axis='y',labelsize=5)

plt.show()
```

### 2010~2017년의 기간동안 서울에서 각 시도로 이동한 인구의 합계를 구하여 그래프로 비교

```python
# 서울에서 충청남도,경상남도, 강원도 전라남도로 이동한 인구 데이터 값만 선택

col_years = list(map(str,range(2010,2018)))
df_4 = df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],col_years]
df_4
```

```python
df_4['합계'] = df_4.sum(axis=1)
df_4
```

```python
df_total =  df_4[['합계']].sort_values(by='합계',ascending=True)
df_total
```

```python
plt.style.use('ggplot')
df_total.plot(kind='barh', # horizontal-bar
              color = 'cornflowerblue',
              figsize = (10,5)
) 
plt.title("서울->타시도 인구 이동")
plt.ylabel('전입지')
plt.xlabel('이동 인구수')
plt.show()
```
