### Deep Learning Model 만들기

```python
from tensorflow import keras
```

```python
model = keras.Sequential()
# 1번째 Hidden Layer : 256 노드 (2의 배수,임의의 개수(Hyper Parameter))
# activation : sigmoid, tanh, relu, softmax, leakyRelu
# 활성화 함수를 사용하지 않게 되면 값의 의미가 없어짐
# 모든 층에는 AF이 있음 
model.add(keras.layers.Dense(256, activation='tanh',input_shape=(3,))) # 은닉층
model.add(keras.layers.Dense(1,activation='sigmoid')) # 출력층
```

```python
# 손실함수
model.compile(optimizer='adam',loss='binary_crossentropy',metrics='accuracy')
# Optimize : learning rate를 알맞게 조정해줌
# - adam 
# - adagrad
# - adaelta
# - rmsprop
# - sgd
# loss : 손실함수
# -binary_crossentropy : 분류 및 확률 예측
# metrics : 모델의 평가 기준
```

```python
# 학습 시키기
from tabnanny import verbose
import numpy as np

model.fit(np.array(x),np.array(y),epochs=500)
```

```python
# 예측하기

model.predict(
    [
        [700,3.7,1],
        [700,3.7,2],
        [700,3.7,3],
        [700,3.7,4],
    ]
)
```
