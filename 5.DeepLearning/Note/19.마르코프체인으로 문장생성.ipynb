### 마르코프 체인으로 문장 생성하기
- 마르코프 체인은 확률을 기반으로 문장을 이어 붙여 나가는 방법
- 워드 샐러드라고도 불림

문장 제작 과정
1) 문장을 단어로 분할(형태소 분석) 합니다.
2) 단어의 전후 연결을 딕셔너리에 등록 합니다.
3) 사전을 이용해 임의의 문장을 생성 합니다.

문장 제작 과정 예
"그는 고양이를 좋아합니다" 
위의 문장을 형태소로 분석하면, "그|는|고양이|를|좋아|합니다"로 분할됨
각 단어의 전후 관계를 알 수 있게 3개의 요소씩 묶어 사전으로 등록한다. 

- 그|는|고양이
- 는|고양이|를
- 고양이|를|좋아
- 를|좋아|합니다

단어와 단어를 처리하고 이를 기반으로 단어들을 연결하여 문자을 생성.    ㅠ

```python
# 텍스트를 한 줄씩 처리하기
twitter = Okt()
lines = text.split("\n")
for line in lines:
    malist = twitter.pos(line,norm=True,stem=True)
    words = []
    for word in malist:
        if not word[1] in ['Punctuation']: # .(마침표)제외한 ? "" ! 를 사용하지 않음
            words.append(word[0])
        if word[0] == ".":
            words.append(word[0])
```

```python
# 딕셔너리 생성
dic = markov.make_dic(words)
json.dump(dic,open(dict_file,"w",encoding="utf-8"))
```
