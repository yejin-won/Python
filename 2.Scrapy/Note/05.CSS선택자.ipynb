```python
from urllib import parse

url = "https://ko.wikisource.org/wiki/저자:윤동주"
# url = "https://ko.wikisource.org/wiki/%EC%A0%80%EC%9E%90:%EC%9C%A4%EB%8F%99%EC%A3%BC"
# browser에 따라 자동으로 바뀔 수 도 있음
url = parse.quote(url)

print(url)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req
from urllib import parse


auth = "저자"
name = "윤동주"

url = "https://ko.wikisource.org/wiki/"
url = url + parse.quote(auth) + ":" + parse.quote(name)

print(url)
```

```python
# urlOpen

res = req.urlopen(url)

# BeautifulSoup으로 분석하기
soup = BeautifulSoup(res,"html.parser")
print(soup)
```

```python
book = soup.select(" div.mw-parser-output > ul a")

for i in range(len(book)):
    if(not book[i].string == "증보판"):
        print("- " + book[i].string)
```
