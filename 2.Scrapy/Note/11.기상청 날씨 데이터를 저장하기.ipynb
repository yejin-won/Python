```python
from bs4 import BeautifulSoup
import requests

# 기상청 주소
req = requests.get("https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108")
html = req.text
soup = BeautifulSoup(html,"html.parser")
```

```python
# 도시출력
for i in soup.find_all("location"):
    print(i.find("city"))
```

```python
# 서울로 테스트
seoul = soup.find("location")
for i in seoul.find_all("data"):
    print(i.find("tmef"))
    print(i.find("wf"))
    print(i.find("tmn"))
    print(i.find("tmx"))
```

```python
# 도시별 날씨 데이터 구축

weather = {}

for i in soup.find_all("location"):
    weather[i.find("city").text] = []
    for j in i.find_all("data"):
        temp = []
        temp.append(j.find("tmef").text)
        temp.append(j.find("wf").text)
        temp.append(j.find("tmn").text)
        temp.append(j.find("tmx").text)
        weather[i.find("city").text].append(temp)
```

```python
print(weather['서울'])
```

```python
print(weather['서귀포'])
```

```python
# 현재 Dictionary로 되어 있어 리스트형태로 변환이 필요
tempList = []
for i in weather:
    for j in weather[i]:
        tempList.append([i,j[0],j[1],j[2],j[3]])
tempList
```

```python
import pandas as pd
```

```python
weatherDataFrame = pd.DataFrame(tempList,columns=['city','tmef','wf','tmn','tmx'])
weatherDataFrame.head()
```

```python
weatherDataFrame.to_csv("../Data/weatherDataFrame.csv",index=False)
pd.read_csv("../Data/weatherDataFrame.csv")
```
