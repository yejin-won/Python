```python
# Library

from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
```

```python
url = "https://movie.daum.net/ranking/boxoffice/weekly"
page= urlopen(url)
soup = BeautifulSoup(page,'html.parser')
```

```python
li_lists = soup.select("#mainContent > div > div.box_boxoffice > ol > li > div > div.thumb_cont > strong > a")
```

```python
# li_lists = soup.select(".link_txt")
# li_lists
# 간단한 버전! => 영화제목에 리스트에 저장할 경우에는 range 조정해야함  
```

```python
count = 0
titleLists = [] # 영화 제목 Lists
countLists = [] # 순위 Lists

for list in li_lists:
    count += 1
    print(count, list.string.strip())
    countLists.append(count)
    titleLists.append(list.string.strip())
```

```python
# Lists 확인
print(countLists)
print(titleLists)
```

```python
# countLists만으로 DataFrame 생성
df = pd.DataFrame(data=countLists,columns=['순위'])
df
```

```python
# 기존 DataFrame에 컬럼과 Lists 추가
df['영화제목'] = pd.Series(titleLists) # Series = vector (in R)
df 
```

```python
# csv로 저장
df.to_csv("../Data/DaumList.csv",encoding='utf-8',index = False)
```

```python
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"
page= urlopen(url)
soup = BeautifulSoup(page,'html.parser')
```

```python
lists = soup.select("#old_content > table > tbody > tr> td.title > div > a")
```

```python
count = 0
countLists = []
titleLists = []

for i in lists:
    count += 1
    countLists.append(count)
    titleLists.append(i.string.strip())
```

```python
df = pd.DataFrame(data=countLists,columns=['Index'])
df
```

```python
df['Titles'] = pd.Series(titleLists)
df
```

```python
df = df[0:5]
```

```python
df.to_csv("../Data/naverRanking.csv",encoding='utf-8',index=False)
```
